```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [2,4,6,2,4,3]
plt.stem(x,y)
print(plt.show())
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [2,4,6,2,4,3]
plt.stem(x,y,linefmt =':', markerfmt ='ro',basefmt = 'g')
print(plt.show())
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [2,4,6,2,4,3]
plt.stem(x,y,linefmt =':', markerfmt ='g',basefmt = 'g')
print(plt.show())
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [2,4,6,2,4,3]
plt.stem(x,y,linefmt =':', markerfmt ='ro',basefmt = 'red',)
print(plt.show())
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [2,4,6,2,4,3]
plt.stem(x,y,linefmt =':', markerfmt ='ro',basefmt = 'red', orientation=False)
print(plt.show())
```
