```python
import matplotlib.pyplot as plt
x =[1,2,3,4,5]
y =[2,4,5,1,3]
plt.plot(x,y)
plt.title('Growth Graph')
plt.text(2,3,'appel', fontsize = 10, style='italic', bbox ={'facecolor':'r'})
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
x =[1,2,3,4,5]
y =[2,4,5,1,3]
plt.plot(x,y)
plt.title('Growth Graph')
plt.text(2,3,'appel', fontsize = 10, style='italic', bbox ={'facecolor':'r'})
plt.annotate('student',xy=(2,1), xytext=(4,4), arrowprops=dict(facecolor='g', shrink = 100))
plt.grid()
plt.show()
```
