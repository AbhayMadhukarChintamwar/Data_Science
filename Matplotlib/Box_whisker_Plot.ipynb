```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70, 130, 150]
x1 = [20,30,40,50,60,70,90,120]
y = [x, x1]

plt.boxplot(y, widths=0.2, labels= ['apple','cherry'], showmeans=True,boxprops=dict(color ='r'), capprops=dict(color ='magenta'),
            whiskerprops=dict(color='aqua'), flierprops=dict(markeredgecolor= 'blue'))
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70, 130, 150]
x1 = [20,30,40,50,60,70,90,120]
x2 = [20,30,40,50,60,70,90,120]
y = [x, x1,x2]

plt.boxplot(y, widths=0.2, labels= ['apple','cherry', 'magoes'], showmeans=True,boxprops=dict(color ='r'), capprops=dict(color ='magenta'),
            whiskerprops=dict(color='aqua'), flierprops=dict(markeredgecolor= 'blue'))
plt.show()
```
