```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [2,4,5,1,3]
plt.step(x, y, color = 'r')
plt.title('Step Plot Graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [2,4,5,1,3]
plt.step(x, y, color = 'r', marker = 'o')
plt.title('Step Plot Graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [2,4,5,1,3]
plt.step(x, y, color = 'r', ms = 10, mfc = 'b')
plt.title('Step Plot Graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [2,4,5,1,3]
plt.step(x, y, color = 'r',marker = 'o')
plt.title('Step Plot Graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend(loc = 4)
plt.grid()
plt.show()
```
