```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [1,2,3,4,5]
plt.plot(x,y, color = 'r')
plt.fill_between(x,y)
plt.title('Fill Between Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,3,4,5])
y = np.array([1,2,3,4,5])
plt.plot(x,y, color = 'r')
plt.fill_between(x,y, where= (x>=2)&(x<=4))
plt.grid()
plt.title('Fill Between Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,3,4,5])
y = np.array([1,2,3,4,5])
plt.plot(x,y, color = 'r')
plt.fill_between(x,y, where= (x>=2)&(x<=4))
# plt.grid()
plt.title('Fill Between Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,3,4,5])
y = np.array([1,2,3,4,5])
plt.plot(x,y, color = 'r')
plt.fill_between(x,y, where= (x>=2)&(x<=4), alpha = 0.4)
plt.title('Fill Between Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```
