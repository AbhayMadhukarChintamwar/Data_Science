```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
var = [1,2,3,4,5,6,7]
var1 = [7,6,5,4,3,2,1]
df = pd.DataFrame({'var':var, 'var1':var1})
sns.lineplot(x = 'var', y = 'var1', data = df)
plt.show()
```

```python
df1 = sns.load_dataset('penguins').head(50)
sns.lineplot(x='bill_length_mm', y = 'flipper_length_mm', data =df1, hue='sex',style='sex',
             palette='rainbow_r', markers=['o','>'], dashes=False)
plt.grid()
plt.show()
```

```python
df1 = sns.load_dataset('penguins').head(50)
sns.lineplot(x='bill_length_mm', y = 'flipper_length_mm', data =df1, hue='sex',style='sex',
             palette='rainbow_r', markers=['o','>'], dashes=False)
plt.grid()
plt.title('Line Graph')
plt.show()
```

```python
df1 = sns.load_dataset('penguins').head(50)
sns.lineplot(x='bill_length_mm', y = 'flipper_length_mm', data =df1, hue='sex',style='sex',
             palette='rainbow_r', markers=['o','>'], dashes=False)
plt.grid()
plt.title('Line Graph')
plt.xlabel('Bill Length')
plt.ylabel('Flipper Length')
plt.show()
```
