```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = sns.load_dataset('penguins')
# print(df)
df
```

```python
sns.scatterplot(x ='bill_length_mm', y ='bill_depth_mm', data =df ,hue ='sex'
                )
plt.show()
```

```python
sns.scatterplot(x ='bill_length_mm', y ='bill_depth_mm', data =df ,hue ='sex',
                style = 'sex')
plt.show()
```

```python
sns.scatterplot(x ='bill_length_mm', y ='bill_depth_mm', data =df ,hue ='sex',
                style = 'sex',size = 'sex', sizes=(60,40))
plt.show()
```

```python
sns.scatterplot(x ='bill_length_mm', y ='bill_depth_mm', data =df ,hue ='sex',
                style = 'sex',size = 'sex', sizes=(60,40), palette= 'rainbow_r')
plt.show()
```

```python
sns.scatterplot(x ='bill_length_mm', y ='bill_depth_mm', data =df ,hue ='sex',
                style = 'sex',size = 'sex', sizes=(60,40), palette='plasma_r')
plt.show()
```
