BAR PLOT IN SEABORN

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
var = [1,2,3,4,5,6,7]
var1 = [7,6,5,4,3,2,1]
df = pd.DataFrame({'var':var, 'var1':var1})
sns.barplot(x = 'var', y = 'var1', data = df)
print(plt.show())
```

```python
df1 = sns.load_dataset('penguins')
sns.barplot(x = df1.island, y = df1.bill_length_mm)
plt.show()
```

```python
sns.barplot(y='bill_length_mm', x= 'island', data = df1, hue='sex')
plt.show()
```

```python
sns.barplot(y='bill_length_mm', x= 'island', data = df1, hue='sex', order=['Dream','Torgersen','Biscoe'])
plt.show()
```

```python
sns.barplot(y='bill_length_mm', x= 'island', data = df1, hue='sex',
            order=['Dream','Torgersen','Biscoe'], hue_order=['Female','Male'])
plt.show()
```

```python
sns.barplot(y='bill_length_mm', x= 'island', data = df1, hue='sex',
            order=['Dream','Torgersen','Biscoe'], hue_order=['Female','Male'], ci =10)
plt.show()
```

```python
sns.barplot(y='bill_length_mm', x= 'island', data = df1, hue='sex',
            order=['Dream','Torgersen','Biscoe'], hue_order=['Female','Male'], ci =10, n_boot=2)
plt.show()
```

```python
sns.barplot(y='bill_length_mm', x= 'island', data = df1, hue='sex',
            order=['Dream','Torgersen','Biscoe'], hue_order=['Female','Male'], palette='rainbow_r', 
            saturation=0.6, errcolor='r')
plt.show()
```

```python
sns.barplot(y='bill_length_mm', x= 'island', data = df1, hue='sex',
            order=['Dream','Torgersen','Biscoe'], hue_order=['Female','Male'], palette='rainbow_r', 
            saturation=0.6, errcolor='r', errwidth=10)
plt.show()
```

```python
sns.barplot(y='bill_length_mm', x= 'island', data = df1, hue='sex',
            order=['Dream','Torgersen','Biscoe'], hue_order=['Female','Male'], palette='rainbow_r', 
            saturation=0.6, errcolor='r', capsize =0.6)
plt.show()
```

```python
sns.barplot(y='bill_length_mm', x= 'island', data = df1, hue='sex',
            order=['Dream','Torgersen','Biscoe'], hue_order=['Female','Male'], palette='rainbow_r', 
          alpha = 0.5,  saturation=0.6, errcolor='r', capsize =0.6)
plt.show()
```

```python
sns.set(style='darkgrid')
sns.barplot(y='bill_length_mm', x= 'island', data = df1, hue='sex',
            order=['Dream','Torgersen','Biscoe'], hue_order=['Female','Male'], palette='rainbow_r',
           errcolor='r', capsize =0.6)

plt.show()
```

```python
sns.set(style='dark')
sns.barplot(y='bill_length_mm', x= 'island', data = df1, hue='sex',
            order=['Dream','Torgersen','Biscoe'], hue_order=['Female','Male'], palette='rainbow_r',
           errcolor='r', capsize =0.6)

plt.show()
```
