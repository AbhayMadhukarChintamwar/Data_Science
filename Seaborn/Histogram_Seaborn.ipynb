HISTOGRAM IN SEABORN

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = sns.load_dataset('penguins')
# print(df)
df
```

```python
sns.displot(x = 'flipper_length_mm',data = df)
plt.show()
```

```python
sns.displot(x = 'flipper_length_mm',data = df, bins=[170, 180, 190,200, 210, 220, 230])
plt.show()
```

```python
sns.displot(x = 'flipper_length_mm',data = df, bins=[170, 180, 190,200, 210, 220, 230],kde=True)
plt.show()
```

```python
sns.displot(x = 'flipper_length_mm',data = df, bins=[170, 180, 190,200, 210, 220, 230],
            kde=True, rug = True)
plt.show()
```

```python
sns.displot(x = 'flipper_length_mm',data = df, bins=[170, 180, 190,200, 210, 220, 230],
            kde=True, rug = True, color='r')
plt.show()
```

```python
sns.displot(x = 'flipper_length_mm',data = df, bins=[170, 180, 190,200, 210, 220, 230],
            kde=True, rug = True,color='gray')
plt.show()
```

```python
sns.displot(x = 'flipper_length_mm',data = df,  color='r', log_scale=True)
plt.show()
```
