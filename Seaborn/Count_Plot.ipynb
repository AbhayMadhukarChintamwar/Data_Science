```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

df = sns.load_dataset('tips')
df
```

```python
sns.barplot(x = 'sex', y = 'total_bill', data = df)
plt.show()
```

```python
sns.countplot(x ='sex', data= df,  palette = 'plasma')
plt.show()
```

```python
sns.countplot(x ='sex', data= df ,hue ='smoker')
plt.show()
```

```python
sns.countplot(y ='sex', data= df ,hue ='smoker')
plt.show()
```

```python
sns.countplot(x ='sex', data= df ,hue ='smoker', color='r')
plt.show()
```

```python
sns.countplot(y ='sex', data= df ,hue ='smoker', color='r')
plt.show()
```

```python
sns.countplot(y ='sex', data= df ,hue ='smoker', palette = 'rainbow_r')
plt.show()
```

```python
sns.countplot(x ='sex', data= df ,hue ='smoker', palette = 'rainbow_r')
plt.show()
```

```python
sns.countplot(x ='sex', data= df ,hue ='smoker', palette = 'rainbow_r', saturation = 0.5)
plt.show()
```
